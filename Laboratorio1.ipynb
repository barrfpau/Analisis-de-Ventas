## **ENERO**

```python
ruta_archivo = '/content/drive/MyDrive/UPSO/BIGDATA/Laboratorio1-Analsis de Ventas Mensuales/Dataset-Labo1/Dataset_deventas_Enero.csv'

df_enero = pd.read_csv(ruta_archivo)
df_enero
```

```python
columnas = df_enero.columns
print(columnas)
```

### Análisis cuantitativo del mes

```python
# Convierto las columnas "Cantidad Pedida" y "Precio Unitario" a tipos numéricos
df_enero["Cantidad Pedida"] = pd.to_numeric(df_enero["Cantidad Pedida"], errors='coerce')
df_enero["Precio Unitario"] = pd.to_numeric(df_enero["Precio Unitario"], errors='coerce')

# Calculo el ingreso total generado en enero
ingreso_total_enero = (df_enero["Cantidad Pedida"] * df_enero["Precio Unitario"]).sum()
print(f"Ingreso total en enero: ${ingreso_total_enero:.2f}")

# Calculo el promedio de ventas diarias en enero
promedio_ventas_diarias_enero = df_enero["Cantidad Pedida"].mean()
print(f"Promedio de ventas diarias en enero: {promedio_ventas_diarias_enero:.2f}")

# Calculo la cantidad total de productos vendidos en enero
cantidad_total_productos_enero = df_enero["Cantidad Pedida"].sum()
print(f"Cantidad total de productos vendidos en enero: {cantidad_total_productos_enero}")

# Utilizo la función describe() para obtener estadísticas básicas
estadisticas_enero = df_enero.describe()
print(estadisticas_enero)
```

### Visualización de las ventas




#### *Producto más vendido del mes*


> Use un gráfico de barras que muestre los productos en el eje X y la cantidad vendida en el eje Y, lo que permitirá identificar claramente cuál fue el producto más vendido en enero.



```python
# Agrupo y cuento la cantidad de cada producto vendido en enero
productos_mas_vendidos = df_enero["Producto"].value_counts()

# Creo un gráfico de barras para mostrar los productos más vendidos
plt.figure(figsize=(10, 5))
ax = productos_mas_vendidos.plot(kind='bar')
plt.title("Productos Más Vendidos en enero")
plt.xlabel("Producto")
plt.ylabel("Cantidad Vendida")

# Roto las etiquetas en el eje X para que sean legibles
ax.set_xticklabels(productos_mas_vendidos.index, rotation=45, ha="right")

plt.grid(True)
plt.tight_layout()  # Ajustar el diseño para evitar recorte de etiquetas
plt.show()

```

#### *Período del mes con más ventas*

Este gráfico de barras muestra los días del mes en el eje X y la cantidad vendida en el eje Y, lo que permitirá identificar en qué días del mes se realizaron más ventas.

```python
# Convierte la columna "Fecha de Pedido" en datetime
df_enero["Fecha de Pedido"] = pd.to_datetime(df_enero["Fecha de Pedido"], errors='coerce')

# Tomo el día del mes de la columna "Fecha de Pedido"
df_enero["Dia del Mes"] = df_enero["Fecha de Pedido"].dt.day

# Agrupo y cuento la cantidad de ventas por día del mes
ventas_por_dia = df_enero.groupby("Dia del Mes")["Cantidad Pedida"].sum()

# Creo un gráfico de barras para mostrar las ventas por día del mes
plt.figure(figsize=(10, 5))
ventas_por_dia.plot(kind='bar')
plt.title("Ventas por Día del Mes en enero")
plt.xlabel("Día del Mes")
plt.ylabel("Cantidad Vendida")
plt.grid(True)
plt.show()

```

#### *Zona con más ventas*

Este gráfico de barras muestra las zonas de envío en el eje X y la cantidad de pedidos en el eje Y, lo que permitirá identificar las zonas desde las cuáles se realizaron más pedidos.

```python

# Agrupo y cuento la cantidad de pedidos por zona de envío
zonas_con_mas_pedidos = df_enero["Dirección de Envio"].value_counts()

# Creo un gráfico de barras para mostrar las zonas con más pedidos
plt.figure(figsize=(10, 5))
zonas_con_mas_pedidos[:10].plot(kind='bar')  # Mostrar las 10 zonas principales
plt.title("Zonas con Más Pedidos en enero")
plt.xlabel("Zona de Envío")
plt.ylabel("Cantidad de Pedidos")
plt.xticks(rotation=45)
plt.grid(True)
plt.show()


```
